```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
%config InlineBackend.figure_formats = ['svg']
```

```python
D=pd.read_csv("../test-data", sep=";", header=None)

grouped = {}
for row in D.iterrows():
    key = row[1][0]
    value = float(row[1][1])
    if not key in grouped:
        grouped[key] = []
    grouped[key].append(value)
```

```python
names = list(grouped.keys())
box_plot_data = list(grouped.values())

fig, ax = plt.subplots(figsize=(8, 6))

#fig.title("Fruit Growth Distribution")
ax.boxplot(box_plot_data, labels=names, bootstrap=None, whis=(0.5,99.5), showfliers=False, vert=0, widths=0.8, sym='.', flierprops=dict(markersize=2))
#ax.set_xlim([-80,0])
plt.plot();
```
